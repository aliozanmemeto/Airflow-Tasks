```python
def fill_missing_dates(product_df, date_col):
  product_df[date_col] = pd.to_datetime(product_df[date_col])
  product_df = product_df.set_index(date_col)

  full_range = pd.date_range(start=product_df.index.min(), end=product_df.index.max(), freq='D')

  return product_df.reindex(full_range, fill_value=0)
```

# Augmented Dickey-Fuller and Kwiatkowski Phillips Schmidt and Shin Test for Stationarity

### We need to reject ADF test and fail to reject KPSS test to indicate stationarity

```python
from statsmodels.tsa.stattools import adfuller, kpss

def test_stationarity(timeseries):
  adf_test = adfuller(timeseries)
  print('ADF Statistic: %f' % adf_test[0])
  print('p-value: %f' % adf_test[1])

  kpss_test = kpss(timeseries)
  print('KPSS Statistic: %f' % kpss_test[0])
  print('p-value: %f' % kpss_test[1])

test_stationarity(product_617453_month['Auftragsmenge'])
test_stationarity(product_617453_month['Auftragsmenge'].diff(12).diff(1).dropna())
```

## From the p value very close to p value 0.05 indicates there is a 12 month seasonality and a trend at lag 1 for the products 617453 and 618868.

```python
test_stationarity(product_618868_month['Auftragsmenge'])
test_stationarity(product_618868_month['Auftragsmenge'].diff(12).diff(1).dropna())
```

## Seasonality per season might also makes sense if we apply differencing by lag 3.

```python
test_stationarity(product_617453_month['Auftragsmenge'])
test_stationarity(product_617453_month['Auftragsmenge'].diff(3).diff(1).dropna())
```

```python
test_stationarity(product_618868_month['Auftragsmenge'])
test_stationarity(product_618868_month['Auftragsmenge'].diff(3).diff(1).dropna())
```

## 617953 already shows stationarity if we only consider ADF test but if we apply log transformation (since there seems to be an increasing variance that violates stationarity) we get more consistent test results.

```python
# Define Metrics

import numpy as np

def mape(actual, forecast):
  """Mean Absolute Percentage Error (MAPE)"""
  actual, forecast = np.array(actual), np.array(forecast)
  return np.mean(np.abs((actual - forecast) / actual))

# MAPE tends to favor models that under-forecast rather than over-forecast. Imagine
# when actual spend value (500) < forecasted spend value (600) , APE is 20.
# However, if actual spend value (600) > forecasted spend value (500) , APE is 16.67 %
def wmape(actual, forecast):
  """Weighted Mean Absolute Percentage Error (wMAPE)"""
  actual, forecast = np.array(actual), np.array(forecast)
  return np.sum(np.abs(actual - forecast)) / np.sum(actual)

# A more logical option in case of sparsity
def smape(actual, forecast):
  """Symmetric Mean Absolute Percentage Error (sMAPE)"""
  actual, forecast = np.array(actual), np.array(forecast)
  denominator = (np.abs(actual) + np.abs(forecast)) / 2
  return np.mean(np.abs(actual - forecast) / denominator)
```

```python
from statsmodels.tsa.arima.model import ARIMA

def fit_arima(df,col, p, d, q, revert_transform = False):
  train_data, test_data = df[0:int(len(df)*0.8)], df[int(len(df)*0.8):]

  train_arima = train_data[col]
  test_arima = test_data[col]

  history = [x for x in train_arima]
  y = [x for x in test_arima]


  # make first prediction
  predictions = list()
  model = ARIMA(history, order=(p, d, q))
  model_fit = model.fit()

  print(model_fit.summary())
  model_fit.plot_diagnostics(figsize=(10,8))
  plt.show()

  yhat = model_fit.forecast(len(test_data))

  if revert_transform:
    train_arima = np.expm1(train_arima)
    y = np.expm1(y)
    yhat = np.expm1(yhat)

  print('Test RMSE: %.3f' % np.sqrt(np.mean((y - yhat)**2)))
  print('Test MAPE: %.3f' % mape(y, yhat))
  print('Test wMAPE: %.3f' % wmape(y, yhat))
  print('Test sMAPE: %.3f' % smape(y, yhat))

  plt.figure(figsize=(16,8))
  plt.plot(train_arima.index, train_arima, color='green', label = 'Train Demand')
  plt.plot(test_data.index, y, color = 'red', label = 'Test Demand')
  plt.plot(test_data.index, yhat, color = 'blue', label = 'Predicted Demand')

  plt.xlabel('Time')
  plt.ylabel('Demand')
  plt.legend()
  plt.grid(True)
  plt.show()
```
